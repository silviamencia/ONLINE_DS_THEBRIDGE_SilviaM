```python
plt.plot(df2['SalePrice'],df2['GrLivArea'],'.')

# Calcular los par√°metros de la regresi√≥n lineal
m, b = np.polyfit(df2['SalePrice'], df2['GrLivArea'], 1)

# Dibujar la recta de regresi√≥n
plt.plot(df2['SalePrice'], m * df2['SalePrice'] + b, color='red')

# A√±adir etiquetas y t√≠tulo si es necesario
plt.xlabel('SalePrice')
plt.ylabel('GrLivArea')
plt.title('Gr√°fico de Dispersi√≥n con Recta de Regresi√≥n')

# Mostrar el gr√°fico
plt.show()
```
