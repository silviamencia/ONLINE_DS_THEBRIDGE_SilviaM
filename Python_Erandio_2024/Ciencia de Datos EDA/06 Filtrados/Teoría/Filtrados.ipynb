Vamos a crear un DataFrame que tiene todas las categorías de "LotShape" excepto "Reg"

Podemos ver cuantos valores de la columna "LotShape" de SalePriceInicios están incluidos en nuestra selección...

### Seleccionar filas de la tabla 1 SI aparecen en la tabla 2.

```python
SalePriceInicios
```

```python
resultado = SalePriceInicios[SalePriceInicios["LotShape"].isin(seleccionados["LotShape"])]
resultado
```

```python
resultado["LotShape"].value_counts()
```

### Seleccionar filas de la tabla 1 SI NO aparecen en la tabla 2.

```python
resultadoNI = SalePriceInicios[~SalePriceInicios["LotShape"].isin(seleccionados["LotShape"])]
resultadoNI2 = SalePriceInicios[SalePriceInicios["LotShape"].isin(seleccionados["LotShape"]) == False]
resultadoNI2 = SalePriceInicios[~SalePriceInicios["LotShape"].isin(seleccionados["LotShape"])]
resultadoNI2
```
