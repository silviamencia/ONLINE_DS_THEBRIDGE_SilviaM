# 5- Realizar dos grupos con observaciones comunes y variables nicas, incluyendo la Id en las dos tablas:


```python
inicio_1 = SalePrice.iloc[0:6, 0:4]

lista = [0] + list(range(7,10))
fin_1 = SalePrice.iloc[4:10, lista]
```

```python
inicio_1
```

```python
fin_1
```

##    a. Incluyendo todas las variables y todas las observaciones, por la Id.


```python
merge_out = pd.merge(inicio_1, fin_1, how='outer', on='Id')
merge_out
```

##    b. Incluyendo las observaciones comunes de las tres tablas y todas las variables, por la Id.


```python
merge_inn = pd.merge(inicio_1, fin_1, on='Id')
merge_inn
```

## c. Incluyendo las observaciones de la tabla dos y enriqueciendo esa informacin con la informacin de la otra tabla, por la Id.


```python
merge_right= pd.merge(inicio_1, fin_1, how='right', on='Id')
merge_right
```

```python
merge_left = pd.merge(inicio_1, fin_1, how='left', on='Id')
merge_left
```
